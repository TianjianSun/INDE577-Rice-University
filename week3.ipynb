```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loss(X, y,w,b):
    M = len(y)
    MSE_loss = sum(.5*(x*w+b-y[i])**2 for i, x in enumerate(X))/M
    return MSE_loss


def gradient_L(X, y, w, b):
    M = len(y)
    partial_w = sum((x*w+b-y[i])*x for i, x in enumerate(X))/M
    partial_b = sum((x*w+b-y[i]) for i, x in enumerate(X))/M
    return partial_w, partial_b


def gradient_descent(X, y, alpha=0.12, max_iter=50):
    random_vec = np.random.rand(2)
    w = random_vec[0]
    b = random_vec[1]
    plt.figure(figsize=(12, 8))
    plt.scatter(X, y)
    colors = {10:'blue', 20:'green', 30:'orange', 50:'red'}
    
    for i in range(1, max_iter+1):
        dw, db = gradient_L(X, y, w, b)
        w = w - alpha * dw
        b = b - alpha * db
        L = loss(X, y, w, b)
        
        if i in [10, 20, 30, 50]:
            print(f'Loss at iter {i} = {L}')
            plt.plot(X, np.array([w*x+b for x in X]), label=f'iter {i}', color = colors[i])
    
    plt.legend(colors)
    plt.show()
```

```python
# class

X = np.array([1,1.5,2.5,4.0])
y = np.array([2,3.5,3,5])

gradient_descent(X, y, alpha=0.02)
```
